### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#for getting urls of the products

URL = []
condition = 0
while condition<3:
    url = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in url:
        URL.append(i.get_attribute("href"))
    
    a = driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    time.sleep(2)
    condition = condition + 1
len(URL)
```

```python
Brand = []
Product_Name= []
Ratings= []
No_of_Ratings= []


for i in URL:
    driver.get(i)
    time.sleep(4)
    
    #Brand
    b = driver.find_element_by_xpath("//div[@class ='a-section a-spacing-none']/a")
    Brand.append(b.text[7:])
    
    #Name of the Product
    n = driver.find_element_by_xpath("//h1[@class ='a-size-large a-spacing-none']")
    Product_Name.append(n.text)
    
    #Ratings
    try:
        r = driver.find_element_by_xpath("//span[@class ='a-size-medium a-color-base']")
        Ratings.append(r.text)
    except NoSuchElementException:
        Ratings.append('-')
   
    #No. of Ratings
    try:
        nr = driver.find_element_by_xpath("//span[@class ='a-size-base a-color-secondary']")
        No_of_Ratings.append(nr.text)
    except NoSuchElementException:
        No_of_Ratings.append('-')
```

```python
Return_Exchange = []
Price= []
Expected_Delivery= []
Availability= []

for i in URL:
    driver.get(i)
    time.sleep(3)
    #Price
    try:
        p = driver.find_element_by_xpath("//span[@id ='priceblock_ourprice']")
        Price.append(p.text)
    except NoSuchElementException:
        p = driver.find_element_by_xpath("//td[@class ='a-span12 a-color-secondary a-size-base']")
        Price.append(p.text) 
    
    #Return/Exchange
    a = driver.find_elements_by_xpath("//a[@class ='a-size-small a-link-normal a-text-normal']")
    for i in range(1,len(a),3):
        Return_Exchange.append(a[i].text)
        
    #Expected delivery
    ed = driver.find_element_by_xpath("//div[@id ='ddmDeliveryMessage']/b")
    Expected_Delivery.append(ed.text)
    
    #Availability
    avail = driver.find_element_by_xpath("//div[@id ='availability']/span")
    Availability.append(avail.text)
```

```python
Seller = []
Color = []
Style_Name = []
Varitation_Configuration = []
Warranty = []

for i in URL:
    driver.get(i)
    time.sleep(4)    
    
    #Seller
    s = driver.find_element_by_xpath("//div[@id ='merchant-info']/a")
    Seller.append(s.text)
    
    #Color
    try:
        c = driver.find_element_by_xpath("//div[@id ='variation_color_name']/div/span")
        Color.append(c.text)
    except NoSuchElementException:
        Color.append('-') 
        
    #StyleName
    try:
        sn = driver.find_element_by_xpath("//div[@id ='variation_style_name']/div/span")
        Style_Name.append(sn.text)
    except NoSuchElementException:
        Style_Name.append('-')
    
    #Varitation_Configuration
    try:
        vc = driver.find_element_by_xpath("//div[@id ='variation_configuration']/div/span")
        Varitation_Configuration.append(vc.text)
    except NoSuchElementException:
        Varitation_Configuration.append('-')
    
    #Warranty
    w = driver.find_elements_by_xpath("//a[@class ='a-size-small a-link-normal a-text-normal']")
    for i in range(3,len(w),3):
        Warranty.append(w[i].text)
```

```python
data = list(zip(Brand,Product_Name,Ratings,No_of_Ratings,Return_Exchange,Price,Expected_Delivery,Availability,Seller,
                Color,Style_Name,Varitation_Configuration,Warranty,URL))


#Creating DataFrame

amazon = pd.DataFrame(data=data, columns = ['Brand','Product_Name','Ratings','No_of_Ratings','Return_Exchange','Price','ExpectedDelivery', 'Availability','Seller','Color','Style_Name','Varitation_Configuration','Warranty','URL'])
amazon.head(4)
```

```python
Up_Vote = []
Comments = []
Time = []

condition=0
while condition<30:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    
    up = driver.find_elements_by_xpath("//span[@id='vote-count-middle']")
    for i in up:
        Up_Vote.append(i.text)
        
    co =  driver.find_elements_by_xpath("//yt-formatted-string[@class ='style-scope ytd-comment-renderer' and @id ='content-text']")
    for i in co:
        Comments.append(i.text.replace('\n',','))
        
    t =  driver.find_elements_by_xpath("//yt-formatted-string[@class ='published-time-text above-comment style-scope ytd-comment-renderer']")
    for i in t:
        Time.append(i.text)
        
    condition = condition + 1
```

```python
Name = []
Distance = []
Ratings = []
Review = []
Dorms = []
Privates = []

condition = True
while condition:
#name
    name = driver.find_elements_by_xpath("//h2[@class='title title-6']")

    for i in name:
        Name.append(i.text)

#distance
    n = driver.find_elements_by_xpath("//span[@class='description']")
    for i in n:
        Distance.append(i.text[-22:].replace(" - ", ''))
    
#Ratings    
    ratings = driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']")
    for i in ratings:
        Ratings.append(i.text[:3].replace('No','-').replace('\n', ''))

#Reviews
    reviews = driver.find_elements_by_xpath("//div[@class='reviews']")
    for i in reviews:
        Review.append(i.text)
    
#for dorm and private

    price = driver.find_elements_by_xpath("//a[@class='prices']/div")[3:]

    for i in range(1,len(price),2):
        Dorms.append(price[i].text[-6:].replace('ilable','-').replace('\n',' ').strip())
    for i in range(0,len(price),2):
        Privates.append(price[i].text[-7:].replace('ailable','-').replace('\n',' ').strip())
    
    try:
        driver.find_element_by_xpath("//div[@class='pagination-item pagination-next']").click()
        time.sleep(2)
    except:
        condition=False
```

```python
Overall_Review = []
Facilities=[]

for i in URL:
    driver.get(i)
    time.sleep(1)    

    #overall
    
    overall = driver.find_elements_by_xpath("//div[@class='keyword']/span")[:1]
    for i in overall:
        Overall_Review.append(i.text)
    
    #facilties

    fa = driver.find_elements_by_xpath("//ul[@class='facility-sections']")
    for i in fa:
        Facilities.append(i.text.replace('\n',','))
```
